# Custom Derivitive



```julia
include("../src/UniversalDiffEq.jl" )
```

```julia
using Random, Lux
dims_in = 2; dims_out = 1; hidden = 10
Random.seed!(1);rng = Random.default_rng() 
NN = Lux.Chain(Lux.Dense(dims_in,hidden,tanh), Lux.Dense(hidden,dims_out))
NNpars, NN_states_model1 = Lux.setup(rng,NN) 
init_params = (NN=NNpars,known_dynamics = (g=2.0, m=0.2, theta=0.3))

function derivs!(du,u,p,t)
    C = NN(u,p.NN,NN_states_model1)[1][1]
    du[1] = p.known_dynamics.g*u[1] - C
    du[2] = p.known_dynamics.theta*C-p.known_dynamics.m*u[2]
end 

model = UniversalDiffEq.CustomDerivatives(data,derivs!,init_params;proc_weight=2.0,obs_weight=0.5,reg_weight = 10.0^-4.5)
nothing
```

```julia
UniversalDiffEq.gradient_decent!(model,verbos = true, maxiter = 250)
```

```julia
UniversalDiffEq.BFGS!(model; verbos = true)
```

```julia
UniversalDiffEq.plot_predictions(model)
```

```julia
plt,(p2,p3)=UniversalDiffEq.plot_forecast(model, 100)
plt
```

```julia
UniversalDiffEq.print_parameter_estimates(model)
```

# Custom differnce equations

```julia
dims_in = 2; dims_out = 1; hidden = 10
Random.seed!(1);rng = Random.default_rng() 
NN = Lux.Chain(Lux.Dense(dims_in,hidden,tanh), Lux.Dense(hidden,dims_out))
NNpars, NN_states = Lux.setup(rng,NN) 
init_params = (NN=NNpars,known_dynamics = (g=2.0, m=0.2, theta=0.3))

function step(u,p)
    C = NN(u,p.NN,NN_states)[1][1]
    du1 = u[1] + p.known_dynamics.g*u[1] - C
    du2 = u[2] -p.known_dynamics.m*u[2] + p.known_dynamics.theta *C
    return [du1,du2] 
end 

model = UniversalDiffEq.CustomDiffernce(data,step,init_params;proc_weight=1.0,obs_weight=1.0,reg_weight = 10^-4)
nothing   
```

```julia
model.loss_function(model.parameters)
```

```julia
UniversalDiffEq.gradient_decent!(model, verbos = true,maxiter = 250)
```

```julia
UniversalDiffEq.BFGS!(model; verbos = true)
```

```julia
UniversalDiffEq.plot_state_estiamtes(model)
```

```julia
UniversalDiffEq.plot_predictions(model)
```

```julia
plt,(p2,p3)=UniversalDiffEq.plot_forecast(model, 100)
plt
```

```julia
UniversalDiffEq.print_parameter_estimates(model)
```
